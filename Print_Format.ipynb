# 基本用法对比

3. f-string格式化：占位符{}，搭配f符号一起使用；

```python
section1= "%s不必说碧绿的菜畦，%s光滑的石井栏，%s高大的皂荚树，%s紫红的桑椹" %(1,2,3,4)
print(section1)
```

```python
section2= "{:d}也不必说鸣蝉在树叶里长吟，{:d}肥胖的黄蜂伏在菜花上，{:d}轻捷的叫天子（云雀）忽然从草间直窜向云霄里去了".format(1,2,3)
print(section2)
```

##  %格式化
1. 填充形式和顺序，只支持固定顺序依次填充
2. 固定长度填充，在字符串左侧填充
3. 对齐，实现对齐功能，只有左对齐和右对齐

```python
# 1. 填充形式和顺序，只支持固定顺序依次填充
test_%s_string1= "%s，%s，%s，%s" %("不必说碧绿的菜畦",2"光滑的石井栏",3"高大的皂荚树",4"紫红的桑椹")
print(test_%s_string1)
```

```python
# 2. 固定长度填充，在字符串左侧填充
# 不指定长度
test_string2 = "%s" %("不必说")
print(test_string2,len(test_string2))
# 指定填充长度为10，字符长度低于指定长度，在左侧填充
test_string3 = "%10s" %("不必说碧绿的菜畦")
print(test_string3,len(test_string3))
#指定填充长度为10，字符长度长于指定长度，填充无效
test_string4 = "%10s" %("不必说碧绿的菜畦，光滑的石井栏")
print(test_string4,len(test_string4))
```

```python
# 3. 对齐，实现对齐功能，只有左对齐和右对齐
# 左填充
test_string5 = '%20s' % ('不必说碧绿的菜畦')
print(test_string5,len(test_string5))
# 右填充
test_string6 = '%-20s' % ('不必说碧绿的菜畦')
print(test_string6,len(test_string6))
```
